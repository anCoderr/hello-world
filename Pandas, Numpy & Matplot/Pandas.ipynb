```python
import pandas as pd
import re
```

# Reading Files

```python
df = pd.read_csv('./pokemon_data.csv')
df_xls = pd.read_excel('./pokemon_data.xlsx')
df_txt = pd.read_csv('./pokemon_data.txt', delimiter='\t') # Tab separated file
df = df[['Name', 'Type 1', 'HP', 'Attack', 'Defense']]
```

# Indexing via Labels: Direct or DF.loc[]

```python
print(df[['Name', 'Attack']])
```

# Indexing via Integer Locations: DF.iloc[]

```python
print(df.iloc[1:4], '\n')
print(df[1:4], '\n')
print(df.iloc[3,1]) # 3rd row 1st column that is 'Name'
```

# Row-Wise Iteration: DF.iterrows()

```python
for index, row in df.iterrows():
    print(index, row)
```

```python
for index, row in df.iterrows():
    print(index, row[['Name', 'Attack']])
```

# Sorting: DF.sort_values

```python
print(df.sort_values(['Type 1', 'HP'], ascending=False), '\n') # Sorts wrt to Type 1 and then HP
print(df.sort_values(['HP', 'Type 1'], ascending=True)) # Sorts wrt to HP and then Type 1
```

```python
print(df.sort_values(['Type 1', 'HP'], ascending=[1, 0])) # Sorts with Type 1 ascending & HP descending
```

# Adding New Columns While Using Some Other Columns:

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense']
print(df.head(5), '\n')
df = df.drop(columns=['Total']) # Doesnt actually alter the df so we reassign df.drop to df
print(df.head(5), '\n')

df['Total'] = df.iloc[:,2:5].sum(axis=1)
print(df.head(5))
```

```python
cols = list(df.columns.values)
cols
```

```python
print(df.head(), '\n')
df = df[cols[0:2] + [cols[-1]] + cols[2:5]]
print(df.head())
```

```python
df.to_csv('modified_with_index.csv')
df.to_csv('modified_without_index.csv', index=False)
```

# Filtering

```python
print(df.loc[df['Type 1'] == 'Fire'][0:4], '\n')
print(df.loc[df['HP'] >= 90][0:4], '\n')
print(df.loc[(df['HP'] >= 20) & (df['Attack'] >= 40)][0:4])
```

```python
print(df.loc[df['Name'].str.contains('Mega')][0:5], '\n')
print(df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)][2:7], '\n') # Case sensitive
print(df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)][2:7]) # Not case sensitive
```

# Conditional Changes

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'FIRE'
print(df[4:7], '\n')
df.loc[df['Type 1'] == 'FIRE', 'Type 1'] = 'Fire'
print(df[4:7])
```

```python
df.loc[df['Total'] > 300, ['Amazing', 'Legend']] = True
print(df[5:10], '\n')
df.loc[df['Total'] > 300, ['Amazing', 'Legend']] = [True, False]
print(df[5:10])
```

# Remove Column: DF.drop()

```python
df.drop(['Amazing', 'Legend'], axis=1)
```

# Get aggregate statistics: DF.groupby()

```python
print(df.groupby(['Type 1']).mean()[0:4])
print(df.groupby(['Type 1']).sum()[0:4])
print(df.groupby(['Type 1']).count()[0:4])
```

```python
df.groupby(['Type 1']).mean().sort_values('Defense')
```

```python
df = df.rename(columns={'Type 1': 'Type'})
print(df)
```
