```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basics of Graphs:

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

x2 = np.arange(0,4.5,0.5)
y2 = x2**2

plt.figure(figsize=(5,3), dpi=150)

plt.plot(x,y, label='2x', color='Red', linewidth=3, linestyle='--', marker='^', markersize=10)
plt.plot(x2[:5], y2[:5], 'go-', label='x^2') # Shorthand notation with [Color][Marker][Line]
plt.plot(x2[4:], y2[4:], 'go--')

plt.title('This is a graph!!')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
gas = pd.read_csv('./gas_prices.csv')
gas
```

```python
# countries = list(gas.columns.values)[1:]
countries =  ['Canada', 'South Korea', 'UK', 'USA']

plt.figure(figsize=(10, 7), dpi=100)

plt.title('Gas Prices Over Time(in USD)')

for country in countries:
    plt.plot(gas['Year'], gas[country], label=country, linewidth=2, linestyle='-', marker='.')

plt.xticks(gas['Year'])

plt.xlabel('Years')
plt.ylabel('US Dollars')

plt.legend()
plt.show()
```

# Basics of Bar Graphs:

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(4,3), dpi=150)

bars = plt.bar(labels, values)

patterns = ['*','/','O']

for bar in bars:
    bar.set_hatch(patterns.pop())

plt.show()
```
